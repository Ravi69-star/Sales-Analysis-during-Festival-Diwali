```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df= pd.read_csv(data, encoding= 'unicode_escape')
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

Now. Lets clean null values

```python
pd.isnull(df)
```

We can see sum of null values in the data.

```python
pd.isnull(df).sum()
```

We will delete sum of null values.

```python
df.dropna(inplace=True)
```

Changing data type from float to int.

```python
df['Amount']= df['Amount'].astype('int')
```

Name of all the columns.

```python
df.columns
```

Plotting single columns using count plot.

```python
 Marital_Status= sns.countplot(data=df, x='Marital_Status')
 for bars in Marital_Status.containers:
  Marital_Status.bar_label(bars)
```

Lets get an idea about which gender has inverted more money.

```python
sales_marital= df.groupby(['Marital_Status'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Marital_Status', y='Amount', data=sales_marital)
```

From above graph we can see the purchasing power of non married is more.

Now we will analyse on the basis of Age Group.

```python
df.columns
```

```python
Age_Group= sns.countplot(data=df, x='Age Group', hue= 'Gender')
for bars in Age_Group.containers:
  Age_Group.bar_label(bars)
```

We can see that the females from age between 26-35 has done more shopping.

Now, FInding the purchasing power on the basis of Age Group.

```python
P_Age_Group= df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(data=P_Age_Group, x= 'Age Group', y='Amount')
```

From Above Bar plot it is clear the purchasing power of Age group between 26-35 is high.

**Analysis with STATE**

Finding the total numbers of orders on the specific state.

```python
df.columns
```

```python
order_state= df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=order_state, x= 'State', y= 'Orders', color='red')
```

We can see that the State 'Uttar Pradesh' has more sales among top 10 states. i.e head(10)

Lets, see the purchasing power of states based on the amount.

```python
Amount_state= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(data=Amount_state, x= 'State', y= 'Amount', color='red')
```

We can out of to 10 states the purchasing power of UP is higher.

**Now, we will do analysis on the basis of Occupation.**

```python
df.columns
```

```python
occupation= sns.countplot(data=df, x='Occupation', color='pink')
for bars in occupation.containers:
  occupation.bar_label(bars)
```

It is confirmed that the most buyrs from IT Sector.

Now, we will find the purchasing power of different Occupations with amount.

```python
P_Occupation_Amount= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=P_Occupation_Amount, x='Occupation', y='Amount')
```

From the above plot, it is clear that the purchasing power of IT sector and Healthcare is at the first.

**Product Category.**

```python
df.columns
```

```python
Products_category=sns.countplot(data=df, x='Product_Category')
sns.set(rc={'figure.figsize': (20, 5)})
for bars in Products_category.containers:
  Products_category.bar_label(bars)
```

From the plot result the highest sales was from Clothing and Apparel.

Now, finding the Purchasing power of different Categories that were sold.

```python
PP_Categories_Amount= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.barplot(data=PP_Categories_Amount, x='Product_Category', y='Amount')
```

The conclusion of data shows that the The highest number of orders were from Clothing and Apparel but most sold product is Food.
